```python
import kagglehub
import pandas as pd

# Download the dataset from KaggleHub
df_path = kagglehub.dataset_download("dineshk125/mall-customer125")

print("Path to dataset files:", df_path)
```

```python
import os

# Construct full path to CSV file (assuming it's named 'Mall_Customers.csv')
file_path = os.path.join(df_path, "Mall_Customers.csv")
df = pd.read_csv(file_path)

df.head()
```

```python
print(df.isnull().sum())
```

```python
df.fillna("Unknown", inplace=True)  # or df.dropna()
```

```python
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(df.dtypes)

# If age or other columns need fixing
df['age'] = df['age'].astype(int)
```

```python
df.to_csv("cleaned_mall_customers.csv", index=False)
```
